```python
import re
from datetime import datetime
import spacy

file_path = "./DL_Text_Extraction_Project/dump/extracted_text.txt"

zip_pattern = r'(\d{5})\b(?!.*\d{5}\b)' # Matches the last 5-digit pattern in a line
date_pattern = r'\d{2}(?:/|-|\\)\d{2}(?:/|-|\\)\d{4}'
dl_pattern = r'\b\w*\d{6,9}\b' # Matches 6 to 9 digits, possibly preceded by letters

zip_codes = []
dates = []
dl_numbers = []

with open(file_path, 'r') as file:
    text = file.read()
    zip_codes = re.findall(zip_pattern, text, flags=re.M)
    dates = re.findall(date_pattern, text)
    dl_numbers = re.findall(dl_pattern, text) # Extracting DL numbers

# spaCy name extraction
nlp = spacy.load("en_core_web_sm")
doc = nlp(text)

names = []
for ent in doc.ents:
    if ent.label_ == "PERSON":
        names.append(ent.text)

for i in range(0, len(zip_codes)):
    date_strs = dates[i * 3:i * 3 + 3]
    # Convert date strings to datetime objects
    date_objs = []
    for date_str in date_strs:
        try:
            separator = re.search(r'[-/\\]', date_str).group()
            date_obj = datetime.strptime(date_str, f"%m{separator}%d{separator}%Y")
            date_objs.append(date_obj)
        except ValueError as e:
            print(f"Warning: Unable to parse date '{date_str}'. Error: {e}")
            continue

    if len(date_objs) != 3:
        print(f"Warning: Incorrect number of dates for Zip Code {zip_codes[i]}. Skipping this license.")
        continue

    # Sort the datetime objects
    date_objs = sorted(date_objs)

    # Assign the dates according to the sorted order
    dob = date_objs[0]
    issue_date = date_objs[1]
    expiration_date = date_objs[2]

    # Print the name, zip code, DL number, and dates
    print(f"\nName: {names[i]}")
    print(f"License with Zip Code: {zip_codes[i]}")
    print(f"Driver's License Number: {dl_numbers[i]}") # Added this line
    print(f"Date of Birth: {dob}")
    print(f"Issue Date: {issue_date}")
    print(f"Expiration Date: {expiration_date}")
```
